```python
file_path = "/content/drive/My Drive/Data/codescie10.txt"

with open(file_path, 'r') as file:
    lines = file.readlines()
pattern = re.compile(r'\s{2,}|\t')
data = [pattern.split(line.strip()) for line in lines]
df = pd.DataFrame(data, columns=['Code', 'Description'])

df['Code'] = df['Code'].str.strip()
df['Description'] = df['Description'].str.strip()
df = df.dropna(subset=['Code', 'Description'])
df = df.reset_index(drop=True)
punto = False
```

```python
def format_codes(code):
  letra = code[0]
  numeros = code[1:]
  if len(numeros) > 2:
    numeros_modificados = numeros[:2] + '.' + numeros[2:]
  else:
    numeros_modificados = numeros
  texto_modificado = letra + numeros_modificados
  return texto_modificado
if not punto:
  df['Code'] = df['Code'].apply(format_codes)
  punto = True
df.tail()
```

```python
df['Description'] = df['Description'].str.lower()
df.tail()
```

```python
df.to_csv('/content/drive/My Drive/Data/codigos_cie_format.csv', index=False)
```

```python
csv_path = '/content/drive/My Drive/Data/DataSetEjemplo_nuestro_statements_cortos.csv'
df = pd.read_csv(csv_path, sep = ';')
```

```python
def transform_codigos(codigos):
    # Separar los códigos por guiones
    codigos_list = codigos.split('-')
    # Añadir un punto después del segundo número en cada código
    codigos_list = [
        codigo[:3] + '.' + codigo[3:] if len(codigo)>3 else codigo
        for codigo in codigos_list
        ]
    # Unir los códigos con comas
    return ','.join(codigos_list)
```

```python
df = df.drop(['CodigosSNOMED'], axis=1)
```

```python
df = df.dropna()
```

```python
df['CodigosCIE_format'] = df['CodigosCIE'].apply(transform_codigos)
df.head()
```

```python
df = df.drop(['CodigosCIE'], axis=1)
```

```python
df = df.rename(columns = {'CodigosCIE_format':'CodigosCIE'})
```
