```python
import pandas as pd
import re


def extract_content_and_metadata(text):
    content_pattern = re.compile(r"page_content='([^']*)'")
    metadata_pattern = re.compile(r"metadata=\{([^}]*)\}")

    contents = content_pattern.findall(text)
    metadatas = metadata_pattern.findall(text)

    # Asegurarse de que la longitud sea 5, rellenando con None si es necesario
    while len(contents) < 5:
        contents.append(None)
    while len(metadatas) < 5:
        metadatas.append(None)

    return contents, metadatas

# Aplicar la funciÃ³n a las columnas 'relevant_chunks' y expandir los resultados en columnas separadas
df_CB[['clean_chunks', 'metadata']] = df_CB['relevant_chunks'].apply(lambda x: pd.Series(extract_content_and_metadata(x)))
df_BB[['clean_chunks', 'metadata']] = df_BB['relevant_chunks'].apply(lambda x: pd.Series(extract_content_and_metadata(x)))
```

# Embeddings para df_BB

```python
df_BB = df_BB.drop(columns=['relevant_chunks'])
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Cargar el tokenizer y el modelo de BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Establecer el dispositivo (GPU si estÃ¡ disponible)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

```

```python
df_BB.info()
```

```python
df_BB = df_BB.drop(columns=['id_note'])
```

```python
# FunciÃ³n para obtener embeddings de BERT
def get_bert_embedding(text):
    if not isinstance(text, str) or text.strip() == "":
        return None  # Devuelve None si la entrada no es vÃ¡lida
    inputs = tokenizer(text, return_tensors='pt', max_length=512, truncation=True, padding=True)
    inputs = {key: value.to(device) for key, value in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
    # Usamos los embeddings de la primera capa oculta (hidden states) del primer token [CLS]
    embeddings = outputs.last_hidden_state[:, 0, :].squeeze().cpu().numpy()
    return embeddings

# Aplicar la funciÃ³n de embeddings a cada celda del dataframe
df_BB_embeddings = df_BB.applymap(lambda cell: get_bert_embedding(cell))

# Visualizar los primeros resultados
df_BB_embeddings.head()
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# FunciÃ³n para calcular la similitud del coseno entre dos vectores
def cosine_similarity_vectors(vec1, vec2):
    vec1 = np.array(vec1).reshape(1, -1)
    vec2 = np.array(vec2).reshape(1, -1)
    return cosine_similarity(vec1, vec2)[0][0]

# FunciÃ³n para reemplazar NaN y None en un vector
def replace_nan_and_none_with_zeros(vec):
    if vec is None:
        return np.zeros(768)
    vec = np.array(vec)
    vec[np.isnan(vec)] = 0
    return vec

# Asegurarse de que los embeddings de 'Notes' estÃ¡n en el formato correcto
note_embeddings = replace_nan_and_none_with_zeros(df_notes_embeddings['Notes'].iloc[0])

# Reemplazar NaN y None en los dataframes
df_BB_embeddings = df_BB_embeddings.applymap(replace_nan_and_none_with_zeros)
df_CB_embeddings = df_CB_embeddings.applymap(replace_nan_and_none_with_zeros)

# Inicializar dataframes para almacenar las similitudes
df_BB_similarities = pd.DataFrame(index=df_BB_embeddings.index, columns=df_BB_embeddings.columns)
df_CB_similarities = pd.DataFrame(index=df_CB_embeddings.index, columns=df_CB_embeddings.columns)

# Calcular similitudes para df_BB_embeddings
for column in df_BB_embeddings.columns:
    df_BB_similarities[column] = df_BB_embeddings[column].apply(lambda x: cosine_similarity_vectors(x, note_embeddings))

# Calcular similitudes para df_CB_embeddings
for column in df_CB_embeddings.columns:
    df_CB_similarities[column] = df_CB_embeddings[column].apply(lambda x: cosine_similarity_vectors(x, note_embeddings))


```
